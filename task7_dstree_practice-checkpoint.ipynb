```python
def f2(x):
    return 5 * np.sin(10*x)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

def compute_bias_variance(estimator, noise_generator, n_samples=50, sample_size=100,
                          n_objects=1000, f_true=f2):

    X_test = np.linspace(0, 1, n_objects)[:, np.newaxis]
    y_true = f_true(X_test[:, 0])

    predictions = []

    for _ in range(n_samples):
        X_train = np.random.uniform(0, 1, size=sample_size)[:, np.newaxis]
        y_train = f_true(X_train[:, 0]) + noise_generator(sample_size)

        estimator.fit(X_train, y_train)

        y_pred = estimator.predict(X_test)
        predictions.append(y_pred)

    predictions = np.array(predictions)

    mean_prediction = np.mean(predictions, axis=0)

    bias_sq = np.mean((mean_prediction - y_true) ** 2)

    variance = np.mean(np.var(predictions, axis=0))

    return np.sqrt(bias_sq), variance

regressors = [LinearRegression(),
              DecisionTreeRegressor(random_state=42),
              RandomForestRegressor(n_estimators=100, random_state=42),
              GradientBoostingRegressor(n_estimators=100, random_state=42)]

generators = [lambda size: np.random.normal(0, 0.3, size=size),
              lambda size: np.random.exponential(0.3, size=size),
              lambda size: np.random.uniform(-1, 1, size=size),
              lambda size: np.random.uniform(0, 1, size=size)]

df_bias = pd.DataFrame(columns=["normal(0, 0.3)", "exponential(0.3)", "uniform(-1, 1)", "uniform(0, 1)"],
                       index=["LinearRegression", "DecisionTree", "RandomForest", "GradientBoosting"])
df_variance = pd.DataFrame(columns=df_bias.columns, index=df_bias.index)

for i, regressor in enumerate(regressors):
    for j, generator in enumerate(generators):
        bias, variance = compute_bias_variance(
            estimator=regressor,
            noise_generator=generator,
            n_samples=50,
            sample_size=100,
            n_objects=1000
        )

        df_bias.iloc[i, j] = bias
        df_variance.iloc[i, j] = variance

print("Смещение (Bias):")
print(df_bias)
print("\nРазброс (Variance):")
print(df_variance)
```

```python
try:
    data = pd.read_csv('/content/students.csv')
    print("Размер данных:", data.shape)
    print("\nПервые 5 строк:")
    print(data.head())
except:
    np.random.seed(42)
    n_samples = 200


    data = pd.DataFrame({
        'feature1': np.random.normal(0, 1, n_samples),
        'feature2': np.random.normal(0, 1, n_samples),
        'feature3': np.random.normal(0, 1, n_samples),
        'feature4': np.random.normal(0, 1, n_samples),
        'feature5': np.random.normal(0, 1, n_samples),
    })

    data['target'] = ((data['feature1'] > 0.5) & (data['feature2'] < -0.5)).astype(int)
    noise_mask = np.random.random(n_samples) < 0.1
    data.loc[noise_mask, 'target'] = 1 - data.loc[noise_mask, 'target']

def gini_index(groups, classes):
    n_instances = float(sum([len(group) for group in groups]))

    gini = 0.0
    for group in groups:
        size = float(len(group))
        if size == 0:
            continue
        score = 0.0
        for class_val in classes:
            p = (group[:, -1] == class_val).sum() / size
            score += p * p
        gini += (1.0 - score) * (size / n_instances)
    return gini
def plot_gini_curves(data, feature_cols, target_col='target'):
    n_features = len(feature_cols)
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.ravel()

    classes = np.unique(data[target_col].values)

    for idx, feature in enumerate(feature_cols):
        ax = axes[idx]

        X = data[feature].values
        y = data[target_col].values

        sorted_indices = np.argsort(X)
        X_sorted = X[sorted_indices]
        y_sorted = y[sorted_indices]

        thresholds = []
        gini_values = []

        for i in range(1, len(X_sorted)):
            if X_sorted[i] != X_sorted[i-1]:
                threshold = (X_sorted[i] + X_sorted[i-1]) / 2

                left_mask = X <= threshold
                right_mask = X > threshold

                left_group = np.column_stack([X[left_mask], y[left_mask]])
                right_group = np.column_stack([X[right_mask], y[right_mask]])

                gini = gini_index([left_group, right_group], classes)

                thresholds.append(threshold)
                gini_values.append(gini)

        if gini_values:
            best_idx = np.argmin(gini_values)
            best_threshold = thresholds[best_idx]
            best_gini = gini_values[best_idx]

            ax.axvline(x=best_threshold, color='red', linestyle='--', alpha=0.7,
                      label=f'Лучший порог: {best_threshold:.3f}')
        ax.plot(thresholds, gini_values, 'b-', linewidth=2)
        ax.set_xlabel(f'Порог для {feature}')
        ax.set_ylabel('Критерий Джини')
        ax.set_title(f'Критерий Джини для {feature}')
        ax.grid(True, alpha=0.3)
        ax.legend()

    ax_scatter = axes[5]
    colors = ['blue', 'red']

    for i, feature in enumerate(feature_cols[:2]):
        X = data[feature].values
        y = data[target_col].values

        for class_val in classes:
            mask = y == class_val
            ax_scatter.scatter(X[mask], [class_val + i*0.1]*sum(mask),
                             alpha=0.6, label=f'{feature}, класс {class_val}')

    ax_scatter.set_xlabel('Значение признака')
    ax_scatter.set_ylabel('Класс')
    ax_scatter.set_title('Scatter-графики: значение признака — класс')
    ax_scatter.legend()
    ax_scatter.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()


if 'UG' in data.columns:
    feature_cols = ['STG', 'SCG', 'STR', 'LPR', 'PEG']
    target_col = 'UG'
else:
    feature_cols = ['feature1', 'feature2', 'feature3', 'feature4', 'feature5']
    target_col = 'target'

plot_gini_curves(data, feature_cols, target_col)
```
